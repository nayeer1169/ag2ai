```python
import os

import autogen
from autogen.agentchat.contrib.math_user_proxy_agent import MathUserProxyAgent

config_list = autogen.config_list_from_json(
    "OAI_CONFIG_LIST.json",
    filter_dict={
        "model": {
            "gpt-4o-mini",
            "gpt-4o",
#            "gpt-35-turbo",
        }
    },
)
from pprint import pprint
pprint(config_list)
```

```python
#in the below code we are firstly creating two agent assistant agent and the mathuser proxy agent
assistant = autogen.AssistantAgent(
    name= "assistant",
    system_message = "You are a helpful assistant", #This sets the system message that instructs the assistant to behave as a helpful entity.
    llm_config={ #Configures the Large Language Model (LLM) with additional settings:
        "timeout": 600,   #Sets the timeout to 600 seconds for any operation.
        "seed":42,  #Sets a seed for random number generation, ensuring reproducibility of results.
        "config_list": config_list,   #This provides the list of configuration data to the assistant for model selection and use.
    },
)

mathproxyagent = MathUserProxyAgent(
    name="mathproxyagent",
    human_input_mode="NEVER",  #this is use so that human cannot give any input
    code_execution_config={"use_docker": False} #use docker is false so that code execution do not use the docker environment
)
```

```python
math_problem = (
    "Find all $x$ that satisfy the inequality $(2x+10)(x+3)<(3x+9)(x+8)$. Express your answer in interval notation."
)

mathproxyagent.initiate_chat(assistant, message=mathproxyagent.message_generator, problem=math_problem)
```

```python
math_problem = "For what negative value of $k$ is there exactly one solution to the system of equations \\begin{align*}\ny &= 2x^2 + kx + 6 \\\\\ny &= -x + 4?\n\\end{align*}"
mathproxyagent.initiate_chat(assistant, message= mathproxyagent.message_generator, problem = math_problem)
```

```python
math_problem = "Problem: If $725x + 727y = 1500$ and $729x+ 731y = 1508$, what is the value of $x - y$ ?"
mathproxyagent.initiate_chat(assistant, message = mathproxyagent.message_generator , problem = math_problem ,prompt_type="python")
```
